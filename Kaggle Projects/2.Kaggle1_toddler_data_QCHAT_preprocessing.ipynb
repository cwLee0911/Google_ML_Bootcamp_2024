```python
# 데이터 불러오기: Toddler Autism dataset July 2018.csv

T_ASD= pd.read_csv('./CH_input/QCHAT_dataset.csv')
T_ASD = T_ASD[['Class/ASD Traits', 'A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10']]
T_ASD = T_ASD.fillna(1)
print("Dataset siez: ", len(T_ASD))
T_ASD.head()
```
